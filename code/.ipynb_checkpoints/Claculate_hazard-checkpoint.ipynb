## Import regression outputs for the 5 specifications, from csv to df

```python
import pandas as pd
import numpy as np
from itertools import product

df = pd.read_csv('../raw_data/output2_clean.csv', sep=',') #output2 is the sample without disability (N=307 055)

df2=df.transpose()
df2.columns = df2.iloc[0]
df2.drop(df2.index[0], inplace=True)
df=df2
df.index=df.index.str.replace('_t:','')
df.index=df.index.str.replace('/:','')
columns_to_convert = ['b1', 'b2', 'b3','b4','b5']  # Specify the columns you want to convert
for column in columns_to_convert:
    df[column] = df[column].astype(float)
df
```

### We only want to keep the model with the least variables (model 1)

```python
model_1=pd.DataFrame({'Model1':df['b1']})
model_1 = model_1.dropna()
model_1 #the constant reports the baseline hazard, gamma is the gompertz parameter, other variables report hazard ratios
```

### Generate hazard function for each age claim and cohort

```python
def get_hazard(in_df, out_df):
    
    #extract the gamma and constant
    gamma = float(in_df.at['gamma','Model1'])
    cons = float(in_df.at['_cons','Model1'])

    #for each of our variable, 
    #extract the first characters of the index to know what level we are considering
    #extract the hazard ratio
    ac_values = [(index[:2], value) for index, value in in_df.loc[in_df.index.str.endswith('.AC'),'Model1'].items()]
    coh_values = [(index[:4], value) for index, value in in_df.loc[in_df.index.str.endswith('.coh'),'Model1'].items()]


    #define the age range we want to generate the hazard rate on
    t = np.arange(-5,46) #from 60 to 110

    #generate all possible combinations of values
    combinations = product(ac_values, coh_values)
    for ac, coh in combinations:
        #compute hazard rate
        fct = cons * ac[1] * coh[1] * np.exp(t * gamma)
        #store the characteristics used in computing and the hazard rate
        new_row = {'AC': ac[0], 'coh': coh[0],'t': t, 'ht': fct}
        #append stored estimates to the out dataframe
        out_df = out_df.append(new_row, ignore_index=True)
    return out_df

#initialize empty dataframe to receive out_df
hazard = pd.DataFrame(columns=['AC', 'coh', 't', 'ht'])
#run the function
hazard = get_hazard(model_1, hazard)
```

### Change the previous table which had the hazard rate saved as a list to a dataframe that has the hazard rate saved as a column with all characteristics used in computing

```python
#initialize receiving df
table=pd.DataFrame()
#multiply all characteristics by 51 so that they are lists like the hazard rate
for var in ['AC','coh']:
    hazard[var] = hazard[var].apply(lambda x: [x] * 51)
    table[var]=hazard[var].explode()
#explode all series into a long form dataframe    
table['t'] = hazard['t'].explode()
table['ht'] = hazard['ht'].explode()
table=table.reset_index()
#column identity serves to indicate separate combinations of characteristics used in computing hazard rate
table=table.rename(columns={'index':'identity'})
table.head(60)
```

```python
def get_survivor(df):
    
    #gen empty column
    St=None
    #loop over rows to compute survival function
    for index, row in df.iterrows():
        t=row['t']
        prop_surv=1-(row['ht'])
        if t==-5:
            St=prop_surv
        else:
            St*=prop_surv
        df.at[index,'St']=St
    return df

table = get_survivor(table)
```
