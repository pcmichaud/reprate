## Compute the hazard rate for $t \in [-5,45]$

$ h(t)=\beta_{0}\prod(\beta_{j})\exp{(\gamma*t)} $, where

$\beta_{0}$ is the constant

$\beta_{j}$ is the hazard ratio of dummy variable j

$\gamma$ is the gompertz parameter

$t$ is the time relative to 65 (so we compute for ages 60 to 110)

### We get hazard rates for ages 60 to 110 for each possible combination of characteristics

### From there, we compute expected lifespan at 60 and 65 for each possible combination of characteristics following the life table methodology
