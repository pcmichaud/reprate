# Function to calculate the present value of a lifetime annuity

## If we assume that the payment is made monthly, we need to adjust the discount rate accordingly. The discount rate needs to be expressed on a monthly basis to match the frequency of payments:
$PV=PMT\frac{1}{r}(1-(1+r)^{-t})$, where $t=e*12$ and $r=r_{month}$

```python
def present_value_lifetime_annuity(payment_per_period, annual_interest_rate, life_expectancy_years):
    """
    Calculate the present value of a lifetime annuity with monthly payments.

    Parameters:
    payment_per_period (float): Payment amount per period (e.g., per month).
    annual_interest_rate (float): Annual interest rate (as a decimal).
    life_expectancy_years (int): Life expectancy in years.

    Returns:
    float: Present value of the lifetime annuity.
    """
    monthly_interest_rate = annual_interest_rate / 12  # Convert annual interest rate to monthly
    total_periods = life_expectancy_years * 12  # Convert life expectancy to months

    if monthly_interest_rate == 0:
        present_value = payment_per_period * total_periods
    else:
        present_value = payment_per_period / monthly_interest_rate * (1 - (1 / (1 + monthly_interest_rate) ** total_periods))
    return present_value
```

```python
# Example usage:
payment_per_month = 10000/12  # Payment per month
annual_interest_rate = 0.05  # Annual interest rate (5%)
life_expectancy_years = 20  # Life expectancy in years
present_value = present_value_lifetime_annuity(payment_per_month, annual_interest_rate, life_expectancy_years)
print("Present Value of Lifetime Annuity (Monthly Payments): {:.2f}$".format(present_value))
```

## The function cumulative_value_cashflow calculates the cumulative value of a given cash flow over a specified age range, taking into account monthly compounding interest. It considers a monthly cash flow amount (cash_flow_per_month), an annual interest rate (annual_interest_rate), and age parameters (start_age and end_age). 

```python
def cumulative_value_cashflow(cash_flow_per_month, annual_interest_rate, start_age, end_age):
    """
    Calculate the cumulative value of a given cash flow over a range of ages with monthly compounding interest.

    Parameters:
    cash_flow_per_month (float): Cash flow amount per month.
    annual_interest_rate (float): Annual interest rate (as a decimal).
    start_age (float): Starting age.
    end_age (float): Ending age.

    Returns:
    float: Cumulative value of the cash flow with interest.
    """
    monthly_interest_rate = annual_interest_rate / 12
    num_months = int((end_age - start_age) * 12)  # Convert age range to months and exclude the end month
    cumulative_value = 0
    for month in range(num_months):
        present_value = cash_flow_per_month * ((1 + monthly_interest_rate) ** month)
        cumulative_value += present_value
    return cumulative_value
```

```python
# Example usage:
cash_flow_per_month = 10000/12  # Cash flow amount per month
annual_interest_rate = 0.05  # Annual interest rate (5%)
start_age = 60  # Starting age
end_age = 65  # Ending age
cumulative_value = cumulative_value_cashflow(cash_flow_per_month, annual_interest_rate, start_age, end_age)
print("Cumulative Value of Cash Flow from age {} to age {}): {:.2f}$".format(start_age, end_age, cumulative_value))
```

## When retirement benefits begin before age 65, the benefit amount is reduced by 0.6% for each month between the beginning of the payments and the 65th birthday. Accordingly, a pensioner who is entitled to the maximum benefit will have the benefit amount reduced by 36% if he or she decides to retire at 60 years of age.

```python
def calculate_reduced_benefit(original_benefit, retirement_age, reduction_rate):
    if retirement_age >= 65:
        return original_benefit  # No reduction if retirement age is 65 or older
    else:
        reduction_percent = (65 - retirement_age) * reduction_rate * 12  # Annual reduction rate
        reduced_benefit = original_benefit * (1 - reduction_percent / 100)
        return reduced_benefit
```

```python
# Example usage:
original_benefit = 100  # Assuming the original benefit amount is $1000
retirement_age = 60  # Age at which retirement benefits begin
reduction_rate = 0.6  # Monthly reduction rate (0.6%)
reduced_amount = calculate_reduced_benefit(original_benefit, retirement_age, reduction_rate)
print("Reduced retirement benefit amount:", reduced_amount)
```

```python
#Example with rrq ajustement
start_age = 60  # Starting age
end_age = 65  # Ending age
adj_factor= 0.6
rrq_65 = 1000
rrq_60 = calculate_reduced_benefit(rrq_65,start_age,adj_factor)
annual_interest_rate = 0.05  # Annual interest rate (5%)
life_expectancy_years = 20  # Life expectancy in years at 65
```

## Value at 65 when you claim at 60

```python
val_claim_60 = cumulative_value_cashflow(rrq_60, annual_interest_rate, start_age, end_age) + present_value_lifetime_annuity(rrq_60, annual_interest_rate, life_expectancy_years)
print("Value when you claim at {}: {:.2f}$".format(start_age,val_claim_60))
```

## Value at 65 when you claim at 65

```python
val_claim_65 = present_value_lifetime_annuity(rrq_65, annual_interest_rate, life_expectancy_years)
print("Value when you claim at {}: {:.2f}$".format(end_age,val_claim_65))
```

## We can create a function that takes in various life expectancies as input, calculates the values at ages 60 and 65 for each life expectancy, and then compares them

```python
def compare_values(rrq_65, annual_interest_rate, start_age, end_age, adj_factor, life_expectancies):
    """
    Compare the values at ages 60 and 65 for different life expectancies.

    Parameters:
    rrq_65 (float): Original RRQ benefit at age 65.
    annual_interest_rate (float): Annual interest rate (as a decimal).
    start_age (int): Starting age for calculations.
    end_age (int): Ending age for calculations.
    adj_factor (float): Adjustment factor representing the reduction rate.
    life_expectancies (list): List of life expectancies to compare.

    Returns:
    dict: Dictionary with life expectancy as key and comparison of values at ages 60 and 65 as value.
    """
    comparisons = {}
    for life_expectancy in life_expectancies:
        # Calculate the reduced RRQ benefit at age 60
        rrq_60 = calculate_reduced_benefit(rrq_65, start_age, adj_factor)

        # Calculate the cumulative value at age 60
        cumulative_value_60 = cumulative_value_cashflow(rrq_60, annual_interest_rate, start_age, end_age) + present_value_lifetime_annuity(rrq_60, annual_interest_rate, life_expectancy)

        # Calculate the present value of a lifetime annuity at age 65
        val_claim_65 = present_value_lifetime_annuity(rrq_65, annual_interest_rate, life_expectancy)

        comparisons[life_expectancy] = {
            "Value at 60": cumulative_value_60,
            "Value at 65": val_claim_65
        }

    return comparisons
```

```python
import matplotlib.pyplot as plt
rrq_65 = 1000
annual_interest_rate = 0.03
start_age = 60
end_age = 65
adj_factor = 0.6
life_expectancies = list(range(5, 30))  # List of life expectancies to compare

comparison_results = compare_values(rrq_65, annual_interest_rate, start_age, end_age, adj_factor, life_expectancies)

life_expectancies = list(comparison_results.keys())
values_at_60 = [comparison_results[life_exp]["Value at 60"] for life_exp in life_expectancies]
values_at_65 = [comparison_results[life_exp]["Value at 65"] for life_exp in life_expectancies]

plt.figure(figsize=(10, 6))
plt.plot(life_expectancies, values_at_60, label='Value at Age 60')
plt.plot(life_expectancies, values_at_65, label='Value at Age 65')
plt.xlabel('Life Expectancy (years)')
plt.ylabel('Value')
plt.title('Comparison of Values at Ages 60 and 65 for Different Life Expectancies')
plt.legend()
plt.grid(True)
plt.show()
```

## We can modify the code to vary the annual interest rate

```python
def compare_values(rrq_65, annual_interest_rates, start_age, end_age, adj_factor, life_expectancies):
    """
    Compare the values at ages 60 and 65 for different life expectancies and interest rates.

    Parameters:
    rrq_65 (float): Original RRQ benefit at age 65.
    annual_interest_rates (list): List of annual interest rates to compare (as decimals).
    start_age (int): Starting age for calculations.
    end_age (int): Ending age for calculations.
    adj_factor (float): Adjustment factor representing the reduction rate.
    life_expectancies (list): List of life expectancies to compare.

    Returns:
    dict: Dictionary with annual interest rate as key and comparison of values at ages 60 and 65 for different life expectancies as value.
    """
    comparisons = {}
    for annual_interest_rate in annual_interest_rates:
        comparison_results = {}
        for life_expectancy in life_expectancies:
            # Calculate the reduced RRQ benefit at age 60
            rrq_60 = calculate_reduced_benefit(rrq_65, start_age, adj_factor)

            # Calculate the cumulative value at age 60
            cumulative_value_60 = cumulative_value_cashflow(rrq_60, annual_interest_rate, start_age, end_age) + present_value_lifetime_annuity(rrq_60, annual_interest_rate, life_expectancy)

            # Calculate the present value of a lifetime annuity at age 65
            val_claim_65 = present_value_lifetime_annuity(rrq_65, annual_interest_rate, life_expectancy)

            comparison_results[life_expectancy] = {
                "Value at 60": cumulative_value_60,
                "Value at 65": val_claim_65
            }
        comparisons[annual_interest_rate] = comparison_results

    return comparisons
```

```python
import matplotlib.pyplot as plt

rrq_65 = 1000
annual_interest_rates = [i / 100 for i in range(4,7)]  # List of annual interest rates from 0% to 7%
start_age = 60
end_age = 65
adj_factor = 0.6
life_expectancies = list(range(5, 21))  # List of life expectancies to compare

comparison_results = compare_values(rrq_65, annual_interest_rates, start_age, end_age, adj_factor, life_expectancies)

plt.figure(figsize=(10, 6))
color_map = plt.cm.get_cmap('tab10', len(annual_interest_rates))  # Get a colormap for the interest rates

for idx, (annual_interest_rate, values) in enumerate(comparison_results.items()):
    color = color_map(idx)
    values_at_60 = [values[life_exp]["Value at 60"] for life_exp in life_expectancies]
    values_at_65 = [values[life_exp]["Value at 65"] for life_exp in life_expectancies]
    plt.plot(life_expectancies, values_at_60, linestyle='--', color=color, label=f'Interest Rate: {annual_interest_rate * 100}% - Age 60')
    plt.plot(life_expectancies, values_at_65, linestyle='-', color=color, label=f'Interest Rate: {annual_interest_rate * 100}% - Age 65')

plt.xlabel('Life Expectancy (years)')
plt.ylabel('Value')
plt.title('Comparison of Values at Ages 60 and 65 for Different Interest Rates and Life Expectancies')
plt.legend()
plt.grid(True)
plt.show()
```

## We modify the code to compare the values at ages 60 and 65 for different adjustment factors and life expectancies

```python
import matplotlib.pyplot as plt

def compare_values(rrq_65, annual_interest_rate, start_age, end_age, adj_factors, life_expectancies):
    """
    Compare the values at ages 60 and 65 for different life expectancies and adjustment factors.

    Parameters:
    rrq_65 (float): Original RRQ benefit at age 65.
    annual_interest_rate (float): Annual interest rate (as a decimal).
    start_age (int): Starting age for calculations.
    end_age (int): Ending age for calculations.
    adj_factors (list): List of adjustment factors (reduction rates) to compare.
    life_expectancies (list): List of life expectancies to compare.

    Returns:
    dict: Dictionary with adjustment factor as key and comparison of values at ages 60 and 65 for different life expectancies as value.
    """
    comparisons = {}
    for adj_factor in adj_factors:
        comparison_results = {}
        for life_expectancy in life_expectancies:
            # Calculate the reduced RRQ benefit at age 60
            rrq_60 = calculate_reduced_benefit(rrq_65, start_age, adj_factor)

            # Calculate the cumulative value at age 60
            cumulative_value_60 = cumulative_value_cashflow(rrq_60, annual_interest_rate, start_age, end_age) + present_value_lifetime_annuity(rrq_60, annual_interest_rate, life_expectancy)

            # Calculate the present value of a lifetime annuity at age 65
            val_claim_65 = present_value_lifetime_annuity(rrq_65, annual_interest_rate, life_expectancy)

            comparison_results[life_expectancy] = {
                "Value at 60": cumulative_value_60,
                "Value at 65": val_claim_65
            }
        comparisons[adj_factor] = comparison_results

    return comparisons
```

```python
rrq_65 = 1000
annual_interest_rate = 0.05
start_age = 60
end_age = 65
adj_factors = [0.5, 0.6, 0.7]  # List of adjustment factors to compare
life_expectancies = list(range(5, 21))  # List of life expectancies to compare

comparison_results = compare_values(rrq_65, annual_interest_rate, start_age, end_age, adj_factors, life_expectancies)

plt.figure(figsize=(10, 6))
values_at_65 = [values[life_exp]["Value at 65"] for life_exp in life_expectancies]
plt.plot(life_expectancies, values_at_65, label='Age 65')
for adj_factor, values in comparison_results.items():
    values_at_60 = [values[life_exp]["Value at 60"] for life_exp in life_expectancies]
    plt.plot(life_expectancies, values_at_60, label=f'Adjustment Factor: {adj_factor} - Age 60',linestyle='--')

plt.xlabel('Life Expectancy (years)')
plt.ylabel('Value')
plt.title('Comparison of Values at Ages 60 and 65 for Different Adjustment Factors and Life Expectancies')
plt.legend()
plt.grid(True)
plt.show()
```
