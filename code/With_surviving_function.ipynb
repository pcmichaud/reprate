### We only want to keep the model with the least variables (model 1)

### Generate hazard function for each age claim and cohort

```python
def get_hazard(in_df, out_df):
    
    #extract the gamma and constant
    gamma = float(in_df.at['gamma','Model1'])
    cons = float(in_df.at['_cons','Model1'])

    #for each of our variable, 
    #extract the first characters of the index to know what level we are considering
    #extract the hazard ratio
    ac_values = [(index[:2], value) for index, value in in_df.loc[in_df.index.str.endswith('.AC'),'Model1'].items()]
    coh_values = [(index[:4], value) for index, value in in_df.loc[in_df.index.str.endswith('.coh'),'Model1'].items()]


    #define the age range we want to generate the hazard rate on
    t = np.arange(0,51) #from 60 to 110

    #generate all possible combinations of values
    combinations = product(ac_values, coh_values)
    for ac, coh in combinations:
        #compute hazard rate
        fct = cons * ac[1] * coh[1] * np.exp(t * gamma)
        #compute survival rate
        st = np.exp((-1*(cons*ac[1]*coh[1])/gamma)*(np.exp(gamma*t)-1))
        #store the characteristics used in computing and the hazard and survival rate
        new_row = {'AC': ac[0], 'coh': coh[0],'t': t, 'ht': fct, 'st':st}
        #append stored estimates to the out dataframe
        print(out_df)
        out_df = out_df.append(new_row, ignore_index=True)
    return out_df

#initialize empty dataframe to receive out_df
hazard = pd.DataFrame(columns=['AC', 'coh', 't', 'ht', 'st'])
#run the function
hazard = get_hazard(model_1, hazard)
```

## Present Value

### Definition

Let $T$ be curtate survival time and $Z$ be the present value of future annuity payments. Denote $v = \frac{1}{1+i}$ and let $B_t$ be the annuity amount to be paid at the end of year $t$.

$Z = \sum_{t=1}^{T} v^t \times B_t$

Then:
    
$E(Z) = \sum_{t=1}^{w} v^{t-1} \cdot B_t \cdot s_t$

Were $w$ is te largest age in the Lifetable

```python
def expected_present_value_future_annuity_payments(B, interest_rate, survival_probabilities):
    """
    Calculate the expected present value of future annuity payments for annuity paid at the beginning of each year.

    Args:
    B (list of float): Annuity amount to be paid at the beginning of each year.
    interest_rate (float): The annual interest rate (expressed as a decimal).
    survival_probabilities (list of float): Probability of surviving in each year.

    Returns:
    float: The expected present value of future annuity payments.
    """
    v = 1 / (1 + interest_rate)
    expected_present_value = 0

    for t in range(len(B)):
        expected_present_value += v**(t) * B[t] * survival_probabilities[t]

    return expected_present_value
```

```python
# Example usage:
B = [0]*5 + [10000] * 46  # Annuity amount to be paid at the end of each year (e.g., for 51 years: 60 to 110 years old)
interest_rate = 0.03  # Annual interest rate (5%)
survival_probabilities = st_id1  # Probability of surviving in each year, here from the Gompertz regression

E_Z = expected_present_value_future_annuity_payments(B, interest_rate, survival_probabilities)
print("Expected Present Value of Future Annuity Payments:", round(E_Z, 2))
```

### The Gompertz mortality law can be represented as:

$\mu(x) = \alpha e^{\beta x}$


where $\mu(x)$  is the mortality rate at age $x$, $\alpha$ is the baseline mortality rate, and $\beta$ is the rate of increase of mortality with age.


Note that since the Gompertz model is for a mortality hazard, we can integrate it
to give us the the survival function:

$h(x) = \alpha e^{\beta x}$ , and

$S(x) = \exp \left( \frac{\alpha}{\beta} (1 - e^{-\beta x}) \right)$

### In stata

$S(t)= \exp\left\{-\lambda_j \gamma^{-1} \left(e^{\gamma t_j} - 1\right)\right\}$, 

$\quad \lambda_j = \exp(x_j \beta)$

$ h(t)=\lambda\exp(\gamma t) $

The model is implemented by parameterizing $\lambda_j = \exp(x_j \beta)$, implying that $h_0(t) = \exp(\gamma t)$, where $\gamma$ is an ancillary parameter to be estimated from the data.

## Reduction for claiming before normal age

### When retirement benefits begin before age 65, the benefit amount is reduced by 0.6% for each month between the beginning of the payments and the 65th birthday. Accordingly, a pensioner who is entitled to the maximum benefit will have the benefit amount reduced by 36% if he or she decides to retire at 60 years of age.

\begin{equation}
\text{reduction_percent} = (65 - \text{retirement_age}) \times \text{reduction_rate} \times 12
\end{equation}

\begin{equation}
\text{reduced_benefit} = \text{original_benefit} \times \left(1 - \frac{\text{reduction_percent}}{100}\right)
\end{equation}

```python
def calculate_reduced_benefit(original_benefit, retirement_age, reduction_rate=0.6):
    if retirement_age >= 65:
        return original_benefit  # No reduction if retirement age is 65 or older
    else:
        reduction_percent = (65 - retirement_age) * reduction_rate * 12 
        reduced_benefit = original_benefit * (1 - reduction_percent / 100)
        return reduced_benefit
```

### Apply reduction rate to the benefits pays at 65

```python
def apply_reduction_to_annuity(B, retirement_age, reduction_rate=0.6):
    return list(map(lambda amount: calculate_reduced_benefit(amount, retirement_age, reduction_rate), B))
```

```python
# Example usage:
B = [10000]*51  # Annuity amount to be paid at the end of each year
retirement_age = 60
reduction_rate = 0.6

reduced_B = apply_reduction_to_annuity(B, retirement_age, reduction_rate)
print("Reduced Annuity Amounts:", reduced_B[0])
```

```python
# Example usage:
B= reduced_B  # Annuity amount to be paid at the end of each year (e.g., $10000 annually for 51 years: 60 to 110 years old)
interest_rate = 0.03  # Annual interest rate (5%)
survival_probabilities = st_id1  # Probability of surviving in each year, here from Gompertz regression

E_Z = expected_present_value_future_annuity_payments(B, interest_rate, survival_probabilities)
print("Expected Present Value of Future Annuity Payments:", round(E_Z, 2))
```

## Comparison of 2 PV

```python
def compare_expected_present_value_annuities(B1, B2, interest_rate, survival_probabilities1, survival_probabilities2):
    """
    Compare the expected present value of two series of annuities.

    Args:
    B1 (list of float): Annuity amounts for series 1.
    B2 (list of float): Annuity amounts for series 2.
    interest_rate (float): The annual interest rate (expressed as a decimal).
    survival_probabilities1 (list of float): Probability of surviving for series 1 in each year.
    survival_probabilities2 (list of float): Probability of surviving for series 2 in each year.

    Returns:
    tuple: A tuple containing the expected present value of series 1 and series 2.
    """
    pv1 = expected_present_value_future_annuity_payments(B1, interest_rate, survival_probabilities1)
    pv2 = expected_present_value_future_annuity_payments(B2, interest_rate, survival_probabilities2)
    return pv1, pv2
```

```python
# Example usage:
B  = [10000]*51
B1 = [0]*5+[10000] * 46  # Annuity amounts for series 1
B2 = apply_reduction_to_annuity(B, retirement_age, reduction_rate)  # Annuity amounts for series 2
interest_rate = 0.03  # Annual interest rate (5%)
survival_probabilities1 = st_id2  # Probability of dying for series 1 in each year
survival_probabilities2 = st_id1  # Probability of dying for series 2 in each year

pv1, pv2 = compare_expected_present_value_annuities(B1, B2, interest_rate, survival_probabilities1, survival_probabilities2)
print("Expected Present Value of Series 1:", round(pv1, 2))
print("Expected Present Value of Series 2:", round(pv2, 2))

if pv1 > pv2:
    print("Series 1 has a higher expected present value.")
elif pv1 < pv2:
    print("Series 2 has a higher expected present value.")
else:
    print("Both series have the same expected present value.")
```

## Progressive annuity

```python
def generate_progressive_annuity(initial_amount, growth_rate, periods):
    """
    Generate a progressive annuity for the first 10 years and remains constant afterward.

    Args:
    initial_amount (float): Initial annuity amount for the first year.
    growth_rate (float): Growth rate of the annuity amount for the first 10 years (expressed as a decimal).
    periods (int): Number of periods (years).

    Returns:
    list: A list containing the annuity amounts for each year.
    """
    annuity_schedule = []
    for t in range(1, periods + 1):
        if t <= 10:
            annuity_schedule.append(initial_amount * (1 + growth_rate) ** (t - 1))
        else:
            annuity_schedule.append(initial_amount * (1 + growth_rate) ** (10 - 1))
    return annuity_schedule
```

```python
# Example usage:
initial_amount = 1000  # Initial annuity amount for the first year
growth_rate = 0.05  # Growth rate of the annuity amount for the first 10 years (5%)
periods = 51  # Number of periods (years)

annuity_schedule = generate_progressive_annuity(initial_amount, growth_rate, periods)
annuity_schedule[0:10]
```

### Comparaison function

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_expected_present_value(initial_amounts, growth_rates, periods, interest_rate, survival_probabilities):
    """
    Plot the expected present value of annuities for different initial amounts and growth rates.

    Args:
    initial_amounts (list of float): Initial annuity amounts for each scenario.
    growth_rates (list of float): Growth rates of the annuity amounts for each scenario (expressed as decimals).
    periods (int): Number of periods (years).
    interest_rate (float): The annual interest rate (expressed as a decimal).
    survival_probabilities (list of float): Probability of surviving for each year.

    Returns:
    None
    """
    grid_growth_rates = np.linspace(growth_rates[0], growth_rates[-1], 100)  # Generate a grid of growth rates

    plt.figure(figsize=(15, 8))
    for initial_amount in initial_amounts:
        expected_present_values = []
        for growth_rate in grid_growth_rates:
            annuity_schedule = generate_progressive_annuity(initial_amount, growth_rate, periods)
            expected_pv = expected_present_value_future_annuity_payments(annuity_schedule, interest_rate, survival_probabilities)
            expected_present_values.append(expected_pv)

        plt.plot(grid_growth_rates*100, expected_present_values, label=f"Initial Amount at 60 : {initial_amount}")

    
    plt.axhline(y=141000, color='r', linestyle='--', label='Horizontal Line at 141K, PV of 6400 claim at 60')
    plt.axvline(x=3.4, color='gray')
    plt.axvline(x=4.9, color='gray')
    plt.xlabel('Growth Rate %')
    plt.ylabel('Present Value')
    plt.title('Present Value of Annuities for Different Initial Amounts and Growth Rates')
    plt.legend()
    plt.grid(True)
    plt.show()
```
